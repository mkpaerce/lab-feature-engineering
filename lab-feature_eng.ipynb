```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check for missing values**

```python
df_missing = spaceship[spaceship.isnull().any(axis=1)]
display(df_missing)
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship_clean = spaceship.dropna()
spaceship_clean.shape
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship_clean['Cabin'].unique()
```

```python
spaceship_clean['Cabin'] = spaceship_clean['Cabin'].str[0]

display(spaceship_clean)
```

```python
spaceship_clean['Cabin'].unique()
```

- Drop PassengerId and Name

```python
spaceship_clean = spaceship_clean.drop(columns = ['PassengerId','Name'], axis=1)
display(spaceship_clean)
```

- For non-numerical columns, do dummies.

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)

ohe.fit(spaceship_clean[['HomePlanet','CryoSleep','Cabin','Destination','VIP','Transported']])
X_train_trans_np = ohe.transform(spaceship_clean[['HomePlanet','CryoSleep','Cabin','Destination','VIP','Transported']])
X_train_trans_np

X_train_trans_df = pd.DataFrame(X_train_trans_np, columns=ohe.get_feature_names_out(), index=spaceship_clean.index) #Need to adjust columns
X_train_trans_df
```

**Perform Train Test Split**

```python
features = X_train_trans_df.drop(columns=['Transported_False','Transported_True'], axis=1)
target = X_train_trans_df['Transported_True']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)

X_train.head()
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
```

- Evaluate your model's performance. Comment it

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
pred
```

```python
y_test.values
```

```python
knn.score(X_test, y_test)

print(f"The accuracy of our model at predicting 'Transported' is {knn.score(X_test, y_test)*100: .2f}%")
```
